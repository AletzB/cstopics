<h1 style="text-align: center;"> Loading Images in Python </h1>

## Importing Libraries

```python
import numpy as np
from scipy.misc import imresize
from imageio import imread
import matplotlib.pyplot as plt
```

## Importing Image

```python
img = imread('./cat.jpg')
plt.imshow(img)
plt.show()
```

## RED Channel

```python
img_red = img * [1.0, 0.0, 0.0]

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Red Image
plt.subplot(1, 2, 2)
plt.imshow(np.uint8(img_red))
plt.show()
```

## GREEN Channel

```python
img_red = img * [0.0, 1.0, 0.0]

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Green Image
plt.subplot(1, 2, 2)
plt.imshow(np.uint8(img_red))
plt.show()
```

## BLUE Channel

```python
img_red = img * [0.0, 0.0, 1.0]

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Blue Image
plt.subplot(1, 2, 2)
plt.imshow(np.uint8(img_red))
plt.show()
```

## Tinted Image

```python
img_tinted = img * [1, 0.9, 0.8]

# Show the original image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Show the tinted image
plt.subplot(1, 2, 2)

# A slight gotcha with imshow is that it might give strange results
# if presented with data that is not uint8. To work around this, we
# explicitly cast the image to uint8 before displaying it.
plt.imshow(np.uint8(img_tinted))
plt.show()
```
