# Loading Images in Python

## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread
import cv2
```

## Importing Image with imageio

```python
img = imread('./cat.jpg')
plt.imshow(img)
plt.show()
```

## RED Channel

```python
img_red = img * [1.0, 0.0, 0.0]

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Red Image
plt.subplot(1, 2, 2)
plt.imshow(np.uint8(img_red))
plt.show()
```

## GREEN Channel

```python
img_red = img * [0.0, 1.0, 0.0]

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Green Image
plt.subplot(1, 2, 2)
plt.imshow(np.uint8(img_red))
plt.show()
```

## BLUE Channel

```python
img_red = img * [0.0, 0.0, 1.0]

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Blue Image
plt.subplot(1, 2, 2)
plt.imshow(np.uint8(img_red))
plt.show()
```

## Tinted Image

```python
img_tinted = img * [1, 0.9, 0.8]

# Show the original image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Show the tinted image
plt.subplot(1, 2, 2)

# A slight gotcha with imshow is that it might give strange results
# if presented with data that is not uint8. To work around this, we
# explicitly cast the image to uint8 before displaying it.
plt.imshow(np.uint8(img_tinted))
plt.show()
```

# Import image with OpenCV

```python
img = cv2.imread('cat.jpg')
plt.imshow(img)
# plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.show()
```

Images in OpenCV are imported in BGR, instead of RGB, that is how matplotlib expect it. To plot it correctly, you can do the following:

```python
img = cv2.imread('cat.jpg')
plt.imshow(img[...,::-1])
plt.show()
```

## Grayscale Images, converting.

```python
img = cv2.imread('cat.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap = 'gray')
plt.show()
```

## Grayscale Images, importing it directly.

```python
img = cv2.imread('cat.jpg', 0)
plt.imshow(img_gray, cmap = 'gray')
plt.show()
```

# Grayscale histogram

```python
img = cv2.imread('cat.jpg',0)
plt.hist(img.ravel(),256,[0,256]); plt.show()
```

# RGB Histogram

```python
img = cv2.imread('cat.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
