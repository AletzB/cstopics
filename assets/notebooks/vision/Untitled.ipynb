```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread
import cv2

maxHist = 3000
```

```python
def linearTrans(img, alpha, beta, dmin=0, dmax=254):
    return np.uint8(np.clip(alpha*img + beta, dmin, dmax))
```

```python
img = imread('./cat.jpg')
plt.subplot(1, 2, 1)
plt.imshow(img)
axes = plt.subplot(1, 2, 2)
plt.hist(img.ravel(),256,[0,256]);
axes.set_ylim([0, maxHist])
plt.show()
```

```python
alpha = 1.0
beta = 70.0
img1 = linearTrans(img, alpha, beta)
plt.subplot(1, 2, 1)
plt.imshow(img1)
axes = plt.subplot(1, 2, 2)
plt.hist(img1.ravel(),256,[0,256]);
axes.set_ylim([0, maxHist])
plt.show()
```

```python
alpha = 1.5
beta = 0.0
img1 = linearTrans(img, alpha, beta)
plt.subplot(1, 2, 1)
plt.imshow(img1)
axes = plt.subplot(1, 2, 2)
plt.hist(img1.ravel(),256,[0,256]);
axes.set_ylim([0, maxHist])
plt.show()
```
